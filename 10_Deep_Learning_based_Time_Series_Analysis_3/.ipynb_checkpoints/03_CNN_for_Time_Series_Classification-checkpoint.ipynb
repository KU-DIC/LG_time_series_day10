```python
import random
import numpy as np
from time import time
import matplotlib.pyplot as plt
#%matplotlib inline #ÏÉùÏÑ±Ìïú figureÎ•º notebookÏóêÏÑú Î≥º ÏàòÏûàÍ≤å Ìï¥Ï£ºÎäî ÏΩîÎìú

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
from torchinfo import summary

#check torch version & device
print ("PyTorch version:[%s]."%(torch.__version__))
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print ("device:[%s]."%(device))
```

```python
# set random seed 

def set_seed(random_seed):
    torch.manual_seed(random_seed)
    torch.cuda.manual_seed(random_seed)
    torch.cuda.manual_seed_all(random_seed)  # if use multi-GPU
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    np.random.seed(random_seed)
    random.seed(random_seed)
    
set_seed(42)
```

```python
# load dataset
from torchvision import datasets,transforms
mnist_train = datasets.FashionMNIST(root='./data/',train=True,transform=transforms.ToTensor(),download=True)
mnist_test = datasets.FashionMNIST(root='./data/',train=False,transform=transforms.ToTensor(),download=True)
# mnist_train = datasets.MNIST(root='./data/',train=True,transform=transforms.ToTensor(),download=True)
# mnist_test = datasets.MNIST(root='./data/',train=False,transform=transforms.ToTensor(),download=True)
print ("mnist_train:\n",mnist_train,"\n")
print ("mnist_test:\n",mnist_test,"\n")
print ("Done.")
```

```python
BATCH_SIZE = 50
train_iter = torch.utils.data.DataLoader(mnist_train,batch_size=BATCH_SIZE,shuffle=True,num_workers=1)
test_iter = torch.utils.data.DataLoader(mnist_test,batch_size=BATCH_SIZE,shuffle=False,num_workers=1)
print ("Load Dataset")
```

### 02. modeling (1) Simple CNN
* Í∞ÑÎã®Ìïú ÌòïÌÉúÏùò CNN Î™®Îç∏ÏùÑ Íµ¨ÌòÑÌï©ÎãàÎã§
* CNN Ï∂úÎ†• ÌÅ¨Í∏∞ Í≥ÑÏÇ∞ÌïòÎäî ÏòàÏãú ÏΩîÎìú (Ï∞∏Í≥†Ïö©) : https://deep-learning-study.tistory.com/471

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(
            in_channels=1,              
            out_channels=16,            
            kernel_size=5,              
            stride=1,                   
            padding=2)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(
            in_channels=16,              
            out_channels=32,            
            kernel_size=5,              
            stride=1,                   
            padding=2)
        self.fc1 = nn.Linear(32 * 7 * 7, 120) # HERE !
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x))) # conv block 1
        x = self.pool(F.relu(self.conv2(x))) # conv block 2
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x) #output
        return x


cnn = SimpleCNN()
loss = nn.CrossEntropyLoss()
optimizer = optim.Adam(cnn.parameters(),lr=1e-3)
print("Done.")
```

```python
summary(cnn, input_size=(50,1,28,28))
```

```python
#training

for epoch in range(5):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(train_iter, 0):
        start_time = time()
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = cnn(inputs) #cnn.forward(inputs)
        loss_out = loss(outputs, labels)
        loss_out.backward()
        optimizer.step()

        # print statistics
        running_loss += loss_out.item()
        if i % 200 == 199:    # print every 200 mini-batches
            print('[{}, {}] loss: {:.3f} time: {:.3f}:'.format
                  (epoch + 1, i + 1, running_loss / 200, (time()-start_time)*200))
            running_loss = 0.0
        
    print(f'End Epoch {epoch+1}')

print('Finished Training')
```

```python
# Testing on CPU

correct = 0
total = 0
with torch.no_grad():
    for data in test_iter:
        images, labels = data
        outputs = cnn(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (
    100 * correct / total))
```

### 03. modeling (2) Complex CNN
* ÎπÑÍµêÏ†Å Ïñ¥Î†§Ïö¥ CNN Î™®Îç∏ÏùÑ Íµ¨ÌòÑÌï©ÎãàÎã§ (Option)

```python
class ConvolutionalNeuralNetworkClass(nn.Module):
    """
        Convolutional Neural Network (CNN) Class
    """
    def __init__(self,name='cnn',input_dim=[1,28,28],
                 kernel_size=3,channel_dims=[32,64],hidden_dims=[256,128],output_dim=10):
        super(ConvolutionalNeuralNetworkClass,self).__init__()
        self.name = name
        self.input_dim = input_dim
        self.kernel_size = kernel_size
        self.channel_dims = channel_dims
        self.hidden_dims = hidden_dims
        self.output_dim = output_dim

        # Convolutional layers
        self.layers = []
        prev_cdim = self.input_dim[0] #input channel = 1 
        for cur_cdim in self.channel_dims: # for each hidden layer
#             print('current channel dimension : ',prev_cdim)
#             print(cur_cdim)
            self.layers.append(
                nn.Conv2d(
                    in_channels=prev_cdim,
                    out_channels=cur_cdim,
                    kernel_size=self.kernel_size,
                    stride=(1,1),
                    padding=self.kernel_size//2
                )) # convolution 
            self.layers.append(nn.ReLU(True))  # activation
            self.layers.append(nn.AvgPool2d(kernel_size=(2,2), stride=(2,2))) # max-pooling 
            self.layers.append(nn.Dropout2d(p=0.5))  # dropout
            prev_cdim = cur_cdim
            

        # Dense layers 
        self.layers.append(nn.Flatten())
        prev_hdim = prev_cdim*(self.input_dim[1]//(2**len(self.channel_dims)))*(self.input_dim[2]//(2**len(self.channel_dims)))
#         print(prev_hdim,self.input_dim[1]//(2**len(self.channel_dims)))
        for cur_hdim in self.hidden_dims:
            self.layers.append(nn.Linear(prev_hdim,cur_hdim,bias=True))
            self.layers.append(nn.ReLU(True))  # activation
            prev_hdim = cur_hdim
                
        # Final layer (without activation)
        self.layers.append(nn.Linear(prev_hdim,self.output_dim,bias=True))

        # Concatenate all layers 
        self.net = nn.Sequential()
        for l_idx,layer in enumerate(self.layers):
            layer_name = "%s_%02d"%(type(layer).__name__.lower(),l_idx)
            self.net.add_module(layer_name,layer)
        self.init_param() # initialize parameters
        
    def init_param(self):
        for m in self.modules():
            if isinstance(m,nn.Conv2d): # init conv
                nn.init.kaiming_normal_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m,nn.Linear): # lnit dense
                nn.init.kaiming_normal_(m.weight)
                nn.init.zeros_(m.bias)
            
    def forward(self,x):
        return self.net(x)

C = ConvolutionalNeuralNetworkClass(
    name='cnn',input_dim=[1,28,28],kernel_size=3,channel_dims=[32,64],
    hidden_dims=[32],output_dim=10).to(device)
loss = nn.CrossEntropyLoss()
optimizer = optim.Adam(C.parameters(),lr=1e-3)
print ("Done.")
```

```python
np.set_printoptions(precision=3)
n_param = 0
for p_idx,(param_name,param) in enumerate(C.named_parameters()):
    if param.requires_grad:
        param_numpy = param.detach().cpu().numpy() # to numpy array 
        n_param += len(param_numpy.reshape(-1))
        print ("[%d] name:[%s] shape:[%s]."%(p_idx,param_name,param_numpy.shape))
        print ("    val:%s"%(param_numpy.reshape(-1)[:5]))
print ("Total number of parameters:[%s]."%(format(n_param,'d')))
```

### 03. Train / Test model

```python
def func_eval(model,data_iter,device):
    with torch.no_grad():
        n_total,n_correct = 0,0
        model.eval() # evaluate (affects DropOut and BN)
        for batch_in,batch_out in data_iter:
            y_trgt = batch_out.to(device)
            model_pred = model(batch_in.view(-1,1,28,28).to(device))
            _,y_pred = torch.max(model_pred.data,1)
            n_correct += (y_pred==y_trgt).sum().item()
            n_total += batch_in.size(0)
        val_accr = (n_correct/n_total)
        model.train() # back to train mode 
    return val_accr
print ("Done")
```

```python
#check eval

C.init_param() # initialize parameters
train_accr = func_eval(C,train_iter,device)
test_accr = func_eval(C,test_iter,device)
print ("train_accr:[%.3f] test_accr:[%.3f]."%(train_accr,test_accr))
```

### Training ! 

```python
start_time = time()

print ("üî•Start trainingüî•")
C.init_param() # initialize parameters
C.train() # to train mode 
epochs,print_every = 3,1
for cur_epoch in range(epochs):
    loss_val_sum = 0
    for batch_in,batch_out in train_iter:
        # Forward path
        y_pred = C.forward(batch_in.view(-1,1,28,28).to(device))
        loss_out = loss(y_pred,batch_out.to(device))
        # Update
        optimizer.zero_grad()      # reset gradient 
        loss_out.backward()      # backpropagate
        optimizer.step()      # optimizer update
        loss_val_sum += loss_out
    loss_val_avg = loss_val_sum/len(train_iter)
    
    # Print
    if ((cur_epoch%print_every)==0) or (cur_epoch==(epochs-1)):
        train_accr = func_eval(C,train_iter,device)
        test_accr = func_eval(C,test_iter,device)
        print ("epoch:[%d] loss:[%.3f] train_accr:[%.3f] test_accr:[%.3f]."%
               (cur_epoch+1,loss_val_avg,train_accr,test_accr))
print ("Training Time (in minutes) = " , (time() - start_time)/ 60)
print ("Done")
```

```python
n_sample = 10
sample_indices = np.random.choice(len(mnist_test.targets),n_sample,replace=False)
test_x = mnist_test.data[sample_indices]
test_y = mnist_test.targets[sample_indices]
with torch.no_grad():
    C.eval() # to evaluation mode 
    y_pred = C.forward(test_x.view(-1,1,28,28).type(torch.float).to(device)/255.)
y_pred = y_pred.argmax(axis=1)
plt.figure(figsize=(10,10))
for idx in range(n_sample):
    plt.subplot(5, 5, idx+1)
    plt.imshow(test_x[idx], cmap='gray')
    plt.axis('off')
    plt.title("Pred:%d, Label:%d"%(y_pred[idx],test_y[idx]))
plt.show()    
print ("Done")
```
